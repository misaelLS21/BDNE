Estas funciones sirven como auxiliares, el verdadero es el que se encuentra un poco más adelante.

```python
os.system('cls')
clear = lambda:os.system('cls')

def menu_inputs():
    """
    @Author:Mls21
    summary: Menú que imprime la interfaz gráfica para ser puente entre Python y Redis
    Parameters: NULL
    Returns: Pantalla de la intefaz gŕafica
    """
    print("")
    print ('+'*20," Datos ", '+'*20, '\n')
    print("0.- Aregar Registro.")
    print("1.- Borrar Registro.")
    print("2.- Ver Registros.")
    print("3.- Salir")




def op5():
    
    while True:
        menu_inputs()
        
        optionM = input("¿Qué desea hacer? >> ")
        #clear()
        
        if optionM == '0': #Agregar Registro
        
            id = input("Ingresa el identificador: ")
            url = input("Ingresa la URL: ")
            nickname = input("Ingrese un nickname al URL: ")
            clasificacion = input("Ingrese la clasificación: ")
            categoria = input("Ingrese catetoria")
        
            #Creamos los conjuntos para cada uno de los datos de los usuarios.
            redis1.sadd('url:'+id,url)
            redis1.sadd('nickname:'+id,nickname)
            redis1.sadd('clasificacion:'+id,clasificacion)
            redis1.sadd('categoria:'+id,categoria)
            
        elif optionM == '1': #Borrar registro
            id = input("Ingresa el identificador: ")
            url = input("Ingresa la URL: ")
            nickname = input("Ingrese un nickname al URL: ")
            clasificacion = input("Ingrese la clasificación: ")
            categoria = input("Ingrese catetoria")
            
            redis1.srem('url:'+id,id)
            redis1.srem('nickname:'+id,nickname)
            redis1.srem('clasificacion:'+id,clasificacion)
            redis1.srem('categoria:'+id,categoria)
            
        elif optionM == '2': #Ver registros.
            id = int(input("Ingresa el identificador: "))
            
            print("Los datos para este usuario son: ")
            sURL = redis1.smembers(f'url:{id}')
            sNickname = redis1.smembers(f'nickname:{id}')
            sClasificacion = redis1.smembers(f'clasificacion:{id}')
            sCategoria = redis1.smembers(f'categoria:{id}')
            
            
            print(f' los url son: {sURL} ') 
            print(f' los nickname son: {sNickname}')      
            print(f' las clasificaciones son: {sClasificacion}')
            print(f' las categorias son: {sCategoria}')
            
        elif optionM == '3':
            break;
            
            
        else:
            print("Opción invalida vuelva a intentarlo.")
            break;
            
            
        input("\n\n Precione ENTER para continuar.")
        clear()

    
        
            
            
            
            
```

Esta aplicación funciona como un menú de una whishlist y tiene un menu conformado 7 opciones, recibe los datos de las url y las guarda datos de la siguiente manera:
<img src="redis5.png" style="width: 300px;">

Cabe mensionar que usa estructuras de datos diferentes:
* Usuarios - Son almacenados en tablas Hashes
* URL, nickname, clasificación y categoria son guardados en conjuntos porque en ellos es fácil poder realizar operaciones clásicas de conjuntos

```python
#funciones auxiliares para ingresar datos por consola en Python.
os.system('cls')
clear = lambda:os.system('cls')

#Obs: La aplicación no acentos comas así que al registrarlas en Redis se deforman.

def menu_aplication():
    """
    @Author:Mls21
    summary: Menú que imprime la interfaz gráfica para ser puente entre Python y Redis
    Parameters: NULL
    Returns: Pantalla de la intefaz gŕafica
    """
    print("")
    print ('+'*20," MENU ", '+'*20, '\n')
    print ("0.- Ingresar nuevo usuario")
    print ("1.- Editar datos de usuario")
    print ("2.- Buscar usuario")
    print ("3.- Mostrar todos los usuarios")
    print ("4.- Cerrar Sesión de un usuario")
    print ("5.- Ingresar registros")
    print ("6.- Comparar registros")
    print ("7.- Salir")
    print ("")

        
# Iniciamos con la interfaz grafica y todo su desarrollo    
while True:
    #inicializamos llamando al menu y eligiendo una opción.
    menu_aplication()
    opMenu = input("Ingrese una opción >> ")
    clear()
    
    if opMenu == '0': #Ingresar nuevo usuario.
        print('-'*5," Ingrese Usuario:", '-'*5, '\n')
        #Ingresar un id para identificar; No. Cuenta para hacerlo práctico.
        id = input("identificador: ")
        #Nombre del usuario.
        usr = input("Usuario: ")
        #Ingresamos una ciudad: Nadamas para hacer bulto
        ciudad = input("Ciudad: ")
        redis1.hmset('usuario:'+id, {'user':usr, 'ciudad':ciudad})
        
    elif opMenu == '1': #Editar datos de usuario
        print('-'*5, "Editar datos de Usuarios", '-'*5, '\n')
        id = input("Ingresa el usuario existente: ")
        #Ingresamos los comandos para ver si es que el usuario existe
        if (redis1.scan(0, 'usuario:'+id)):
            usr = input("Nuevo usuario: ")
            ciudad = input("Nueva ciudad")
            redis.hmset('usuario:'+id, {'user':usr, 'ciudad':ciudad})
        else:
            print('El usuario no existe')
            
    elif opMenu == '2': #Buscar Usuario.
        print('-'*10, 'Buscador', '-'*10, '\n')
        id = input('Ingrese identificador: ')
        print(redis1.hgetall('usuario:'+id), '\n')
        print('Usuario: ', redis1.hget('usuario:'+id, 'user').decode())
        print('Ciudad: ', redis1.hget('usuario:'+id, 'ciudad').decode())
        
    elif opMenu == '3': #Mostrat todos los usuarios.
        print('-'*10, 'Usuarios Registrados', '-'*10, '\n')
        #Aquí hay que ver si podemos hacer algo para corregir el formato.
        print("Hay ", redis1.scan(0, 'usuario:*'), " usuarios autenticados") 
        
    elif opMenu == '4': #Cerrar sesión de usuario
        print('-'*5, "Cerrar Sesión de un usuario", '-'*5, '\n')
        id = input("Identificador: ")
        if (redis1.scan(0, 'usuario:'+ id)):
            redis1.delete('usuario:' + id) #borramos todos los urls de este usuario y así sucesivamente.
            redis1.delete('url:'+id)
            redis1.delete('nickname:'+id)
            redis1.delete('categoria:'+id)
            redis1.delete('clasificacion:'+id)
        else: 
            print("El usuario no existe")
    
    elif opMenu == '5': #Ingresar registros
        #Iniciamos a trabajar con los registros los cuales los guardaremos en sets porque son manipulables
        op5()
        
    elif opMenu == '6': #Comparamos los registros para encontrar matches comunes
        usuario1 = input("Ingresa el identificador del usuario1: ")
        usuario2 = input("Ingresa el identificador del usuario2: ")
        #Creamos un nuevo conjunto en el cual guardaremos los datos de la intersección
        int_dato = 'interseccion' + usuario1 + usuario2
        
        #Definimos los conjuntos delas intersecciones
        url_inter = redis1.sinterstore('url'+int_dato,'url:'+usuario1,'url:'+usuario2)
        nick_inter = redis1.sinterstore('nickname'+int_dato, 'nickname:'+usuario1, 'nickname:'+usuario2)
        cate_inter = redis1.sinterstore('categoria'+int_dato, 'categoria:'+usuario1, 'categoria:'+usuario2)
        clasif_inter = redis1.sinterstore('clasificacion'+int_dato, 'clasificacion:'+usuario1, 'clasificacion:' + usuario2)
        
        #Mandamos a llamar el nuevo conjunto donde se almacenaron las interseciones
        sURL = redis1.smembers(f'url{int_dato}')
        sNickname = redis1.smembers(f'nickname{int_dato}')
        sClasificacion = redis1.smembers(f'clasificacion{int_dato}')
        sCategoria = redis1.smembers(f'categoria{int_dato}')
            
        #Imprimimos los registros comunes para todos los registros registrados por cada uno de los usuarios    
        print(f' los urls en comunes son: {sURL} ') 
        print(f' los nicknames comunes son: {sNickname}')      
        print(f' las clasificaciones comunes son: {sClasificacion}')
        print(f' las categorias comunes son: {sCategoria}')
        
        
        
    
    else:
        break;
    
    input("\n\n Precione ENTER para continuar.")
    clear()
        
        
    
    
    

    
    
    
```
